```python
import pandas as pd
import numpy as np
import json
```

```python
movies=pd.read_csv('movies.csv')
```

```python
credits=pd.read_csv('credits.csv')
```

```python
movies
```

```python
credits
```

```python
movie_credits=pd.merge(movies,credits,left_on='id',right_on='movie_id')
```

```python
movie_credits=movie_credits[['keywords','genres','original_title','overview','cast','crew','movie_id']]
```

```python
movie_credits=movie_credits.dropna()
```

```python
movie_credits
```

```python
def fetch_director(value):
    values = json.loads(value)
    return [i['name']for i in values if i['job'] =='Director']
```

```python
fetch_director(movie_credits['crew'][0])
```

```python
movie_credits['crew']=movie_credits['crew'].apply(fetch_director)
```

```python
def fetch_cast(value):
    values = json.loads(value)
    return [i['name']for i in values][:3]
```

```python
fetch_cast(movie_credits['cast'][0])
```

```python
movie_credits['cast']=movie_credits['cast'].apply(fetch_cast)
```

```python
def extract(value):
    values = json.loads(value)
    return [i['name']for i in values]
```

```python
movie_credits
```

```python
movie_credits['genres']=movie_credits['genres'].apply(extract)
movie_credits['keywords']=movie_credits['keywords'].apply(extract)
```

```python
movie_credits
```

```python
movie_credits['overview']=movie_credits['overview'].str.split()
```

```python
movie_credits['overview']
```

```python
movie_credits
```

```python
def collapse(value):
    final_list =[]
    for i in value:
        final_list.append(i.replace(' ',''))
    return final_list
```

```python
collapse(movie_credits['cast'][0])
```

```python
movie_credits['cast']=movie_credits['cast'].apply(collapse)
movie_credits['crew']=movie_credits['crew'].apply(collapse)
movie_credits['genres']=movie_credits['genres'].apply(collapse)
movie_credits['keywords']=movie_credits['keywords'].apply(collapse)
```

```python
movie_credits
```

```python
movie_credits['tags'] = movie_credits['cast'] + movie_credits['keywords'] + movie_credits['crew'] + movie_credits['genres'] + movie_credits['overview']
```

```python
movie_credits.drop(columns=['keywords','genres','overview','cast','crew'],axis=1,inplace=True)
```

```python
movie_credits.rename({'original_title':'title'},axis=1,inplace=True)
```

```python
movie_credits
```

```python
movie_credits['tags']=movie_credits['tags'].apply(lambda x:" ".join(x))
```

```python
movie_credits
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(value):
    lst = []
    for i in value.split():
        lst.append(ps.stem(i))
    return " ".join(lst)
```

```python
stem(movie_credits['tags'][0])
```

```python
movie_credits['tags']=movie_credits['tags'].apply(stem)
```

```python
movie_credits
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(movie_credits['tags'])
```

```python
vector = cv.fit_transform(movie_credits['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

```python
similarity[0]
```

```python
movie_index = movie_credits[movie_credits['title'] == 'The Dark Knight Rises'].index[0]
movie_index
```

```python
distances = similarity[movie_index]
```

```python
distances
```

```python
index_list=list(enumerate(distances))
index_list[:5]
```

```python
similar_movie=sorted(index_list , reverse = True , key= lambda x: x[1])[:11]
similar_movie
```

```python
for ind, i in similar_movie:
    print(movie_credits.iloc[ind,0])
```

```python
def recommender(movie_name):
    movie_index = movie_credits[movie_credits['title'] == movie_name].index[0]
    distances = similarity[movie_index]
    index_list = list(enumerate(distances))
    similar_movie = sorted(index_list, reverse=True, key=lambda x: x[1])[1:11]
    for ind , i in similar_movie:
        print(movie_credits.iloc[ind,0])
```

```python
recommender('Iron Man')
```
